```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a synthetic dataset
data = {
    'age': [25, np.nan, 35, 45, 50, 28, 34, 46, 55, np.nan],
    'salary': [50000, 60000, np.nan, 80000, 85000, 48000, 52000, 82000, 90000, 75000],
    'city': ['New York', 'Los Angeles', 'New York', 'San Francisco', 'Los Angeles', 'San Francisco', 'New York', 'San Francisco', 'Los Angeles', 'New York'],
    'purchased': ['No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes']
}
```

```python
df = pd.DataFrame(data)
```

```python
# Handling missing values for visualization purposes (imputation with mean for simplicity)
df['age'].fillna(df['age'].mean(), inplace=True)
df['salary'].fillna(df['salary'].mean(), inplace=True)
```

```python
# Plotting histograms
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
sns.histplot(df['age'], kde=True, bins=10)
plt.title('Distribution of Age')
```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['salary'], kde=True, bins=10)
plt.title('Distribution of Salary')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Plotting scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='salary', hue='purchased', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Plotting box plots
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x='city', y='age', data=df)
plt.title('Age Distribution by City')
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(x='city', y='salary', data=df)
plt.title('Salary Distribution by City')
plt.tight_layout()
plt.show()
```
